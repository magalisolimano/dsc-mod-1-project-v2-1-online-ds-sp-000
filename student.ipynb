```python
# Convert foreign_gross to float; Rename col and create new worldwide col
df1['foreign_gross'] = pd.to_numeric(df1['foreign_gross'], errors='coerce')
display(df1.info())
df1.head()
```

```python
# Convert budget and revenue cols to float dtype; release_date to datetime. Remove '$' and ','.
df11['production_budget'] = df11['production_budget'].map(lambda x: float(x.replace("$", "").replace(',','')))
df11['domestic_gross'] = df11['domestic_gross'].map(lambda x: float(x.replace("$", "").replace(',','')))
df11['worldwide_gross'] = df11['worldwide_gross'].map(lambda x: float(x.replace("$", "").replace(',','')))
df11['release_date'] = pd.to_datetime(df11['release_date'])
df11['year'] = pd.DatetimeIndex(df11['release_date']).year

# Confirm that dtype conversions worked and that null values did not increase
display(df11.info())
```

```python
df11 = df11.rename(columns={"movie": "title"})
df11.head()
```

```python
df_revenue = df11.join(df1, how='left', rsuffix='_right')
display(df_revenue.info())
df_revenue.head(20)
```

```python
# Drop df1 revenue data since they seem to round, while df11's revenue data seems more precise.
# Drop foreign_gross, year_right.
```

```python
df_revenue = df_revenue.drop(columns=['domestic_gross_right', 'foreign_gross', 'year_right'])
```

```python
# Calculate foreign gross revenue as difference between worldwide and domestic revenue
df_revenue['foreign_gross'] = df_revenue['worldwide_gross'] - df_revenue['domestic_gross']
df_revenue.head()
```

```python
# Convert "0" values to NaN
df_revenue['domestic_gross'] = df_revenue['domestic_gross'].replace(0.0, np.nan)
df_revenue['worldwide_gross'] = df_revenue['worldwide_gross'].replace(0.0, np.nan)
df_revenue['foreign_gross'] = df_revenue['foreign_gross'].replace(0.0, np.nan)
df_revenue.head(10)
```

```python
df_revenue['worldwide_net_revenue'] = df_revenue['worldwide_gross'] - df_revenue['production_budget']
df_revenue
```

```python
display(df_revenue.info())
df_revenue.apply(['mean', 'median', 'std'])
```

```python
# Domestic and worldwide gross revenue, net revenue are positively skewed--mean is larger than median, with large outliers distorting the mean.
# Explore distribution of worldwide_net_revenue
fig, ax = plt.subplots(figsize=(6,6))
ax.hist(df_revenue['worldwide_net_revenue'], bins='auto', color = 'green', alpha = 0.5)
ax.set_title('Distribution of Worldwide Net Revenue')
ax.set_xlabel('Net Revenue ($ Billions)')
ax.set_ylabel('Number of Movies')
ax.axvline(df_revenue['worldwide_net_revenue'].mean(), color = 'black')
```
